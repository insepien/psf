```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#d = pickle.load(open("dict2pR/dict_rand_7.pkl", "rb"))
```

```python
seeds = list(range(6,11))+[22,23,25,26,27]+list(range(30,35))+list(range(36,40))
seeds +=list(range(42,54))+list(range(55,61))+list(range(62,70))+list(range(71,91))
seeds += list(range(90,123))+list(range(124,130))
```

```python

dirr = ["dict2p", "dict2pR"]
def get_e(dictdir, kind):
    e1 = {}
    e2 = {}
    sigma ={}
    label = ['e1','e2','sigma']
    item = [e1,e2,sigma]
    for j in seeds:
        data = pickle.load(open(dictdir+"/dict_"+kind+"_"+str(j)+".pkl", "rb"))
        for i in range(3):
            item[i][j] = [data[label[i]]['g1'][0],data[label[i]]['g2'][0]]

    e1 = pd.DataFrame(e1, index=["e1_g1","e1_g2"]).T
    e2 = pd.DataFrame(e2, index=["e2_g1","e2_g2"]).T
    #sigma = pd.DataFrame(sigma, index=["sigma_g1","sigma_g2"]).T
    e = pd.concat([e1,e2], axis=1)
    return e

eP = get_e(dirr[0], "psfws")
eR = get_e(dirr[1], "rand")
eP.set_index([pd.Index(list(range(len(eP))))], inplace=True)
eR.set_index([pd.Index(list(range(len(eP))))], inplace=True)
```

```python
#MY MATH
def get_phi(g1, g2):
    # this give theta in degrees
    phi = np.arctan2(g2,g1)*180/np.pi
    for i in range(len(g1)):
        if g1[i]*g2[i] > 0:
            #first, quadrant
            if g1[i] > 0:
                phi[i] = phi[i]*0.5
            #third quadrant
            else: 
                phi[i] = (360+phi[i])/2
        # second quadrant
        elif g1[i] < 0 and g2[i] > 0:
            phi[i] = (phi[i] +180)/2
        # fouth quadrant
        elif g1[i] > 0 and g2[i] < 0:
            phi[i] = (phi[i]+180)/2  
    return phi

eP["e1_phiP"] = get_phi(eP["e1_g2"],eP["e1_g1"])
eP['e2_phiP'] = get_phi(eP["e2_g2"],eP["e2_g1"])
eR['e1_phiR'] = get_phi(eR["e1_g2"],eR["e1_g1"])
eR['e2_phiR'] = get_phi(eR["e2_g2"],eR["e2_g1"])
```

```python
plt.rcParams['font.size'] = 17
plt.rcParams['axes.labelsize'] = 17
things = [eP["e1_phiP"], eP['e2_phiP'], eR['e1_phiR'], eR['e2_phiR']]
fig,ax =plt.subplots(2,2,figsize=(12,7), sharey=True, sharex=True)
binn = 25
y, bin_edges, _ = ax[0,0].hist(things[0], bins=binn, histtype='step', color="white")
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
ax[0,0].errorbar(
    bin_centers,
    y,
    yerr = y**0.5,
    marker = '.', drawstyle="steps-mid",
    linestyle ="",capsize=3, color='blue'
)
y, bin_edges, _ =ax[0,1].hist(things[1], bins=binn, histtype='step', color="white")
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
ax[0,1].errorbar(
    bin_centers,
    y,
    yerr = y**0.5,
    marker = '.',
    drawstyle = 'steps-mid',
    linestyle ="",capsize=3, color="blue"
)
y, bin_edges, _ =ax[1,0].hist(things[2], bins=binn, histtype='step', color="white")
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
ax[1,0].errorbar(
    bin_centers,
    y,
    yerr = y**0.5,
    marker = '.',
    drawstyle = 'steps-mid',
    linestyle ="",capsize=3, color="magenta"
)
y, bin_edges, _ =ax[1,1].hist(things[3], bins=binn, histtype='step', color="white")
bin_centers = 0.5*(bin_edges[1:] + bin_edges[:-1])
ax[1,1].errorbar(
    bin_centers,
    y,
    yerr = y**0.5,
    marker = '.',
    drawstyle = 'steps-mid',
    linestyle ="",capsize=3, color="magenta"
)
ax[0,0].set_title("e1, psf-weather-station")
ax[0,1].set_title("e2, psf-weather-station")
ax[1,0].set_title("e1, random")
ax[1,1].set_title("e2, random")

[a.set_xlabel("Correlation function angle", labelpad=10) for a in ax[1]]
ax[0,0].set_ylabel("Frequency")
ax[1,0].set_ylabel("Frequency")
ax[0,0].set_xticks(np.linspace(0,180,5))

fig.tight_layout()
```
