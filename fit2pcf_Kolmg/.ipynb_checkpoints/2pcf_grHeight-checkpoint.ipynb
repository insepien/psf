# notes
- this notebook plots the correlation func of sigma, e1,e2 as function of separation
- same heights.
- pkl dir: heightPsfws, heightRand

```python
import psfws
import pickle
import numpy as np
import treecorr
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.axes_grid1 import make_axes_locatable
import treecorr
import pandas as pd
import galsim

plt.rcParams['font.size'] = 15
plt.rcParams['axes.labelsize'] = 15
```

```python
def corr1d(thx,thy,para):
    cat = treecorr.Catalog(x=thx, y=thy, k=para, x_units='degree', y_units='degree')
    kk = treecorr.KKCorrelation(min_sep=0.001, max_sep=1, bin_type="Log",nbins=20, sep_units='degree')
    kk.process(cat)
    xi = kk.xi
    r = kk.rnom
    return xi, r
```

## correlation functions for random-parameter simulations with different 0th layer height
For different h0, random sims h0 = 0.1m
For same h0, random sims h0 = 0.2 m

```python
meansRdh={}
meansRsh={}

def get_corrs(path, i):
    data = pickle.load(open(path+str(i)+".pkl", 'rb'))
    for key in data.keys():
        data[key] = np.array(data[key])
    (thx, thy, seed, x, y, sigma, e1, e2, arguments, atmSummary, atmKwargs) = tuple(data.values())

    corrSigma,xs = corr1d(thx,thy,sigma)
    corre1,xs = corr1d(thx,thy,e1)
    corre2,xs = corr1d(thx,thy,e2)
    return  corrSigma, corre1, corre2, xs

for i in ([6, 7, 8, 9,10, 22, 23, 25, 26, 27]):
    corrsigmaDH, corre1DH, corre2DH, xs = get_corrs("../runsims/diffheightRand/out_rand_", i)
    corrsigmaSH, corre1SH, corre2SH, xs = get_corrs("../runsims/sameheightRand/outh_rand_", i)
    
    fig, ax = plt.subplots(1,3,figsize=(13,3))
    
    paraDH = [corrsigmaDH, corre1DH, corre2DH]
    paraSH = [corrsigmaSH, corre1SH, corre2SH]
    
    [ax[i].scatter(xs, paraDH[i], c="b", label="h0=0.1") for i in range(3)]
    [ax[i].scatter(xs, paraSH[i], c="orange", label="h0=0.2") for i in range(3)]
    
    lbl = ["$\\xi_{sigma}$", "$\\xi_{e1}$", "$\\xi_{e2}$"]
    [ax[i].set_ylabel(lbl[i]) for i in range(3)]
    [ax[i].set_xlabel("$\delta(r)$") for i in range(3)]
    
    [a.ticklabel_format(style="sci", scilimits=(0,0)) for a in ax]
    [a.set_xscale('log') for a in ax]
    
    fig.suptitle("seed " + str(i))
    
    if i==6 or i==8:
        fig.legend(loc='center left', bbox_to_anchor=(1, 0.5))
        
    meansRdh[i] = [np.mean(corrsigmaDH), np.mean(corre1DH), np.mean(corre2DH)]
    meansRsh[i] = [np.mean(corrsigmaSH), np.mean(corre1SH), np.mean(corre2SH)]
    
    
    if i not in list(range(6,11)):
        plt.close()
    fig.tight_layout()
    
```

# plot mean corr vs mean speed for psfws para

```python
# mean 
meanspd = []
corr2m = {}
corr4m = {}


for i in ([6, 7, 8, 9,10, 22, 23, 25, 26, 27]):
    #2m
    data = pickle.load(open("../runsims/sameheightRand/outh_rand_"+str(i)+".pkl", 'rb'))
    (thx, thy, seed, x, y, sigma, e1, e2, arguments, atmSummary, atmKwargs) = tuple(data.values())
    
    meanspd.append(np.ceil(np.mean(atmKwargs["speed"])))
    corrs_2m,_ = corr1d(thx,thy,sigma)
    corre1_2m,_ = corr1d(thx,thy,e1)
    corre2_2m,_ = corr1d(thx,thy,e2) 
    corr2m[i] = [np.mean(corrs_2m), np.mean(corre1_2m), np.mean(corre2_2m)]
    
    #4m
    data = pickle.load(open("../runsims/diffheightRand/out_rand_"+str(i)+".pkl", 'rb'))
    (thx, thy, seed, x, y, sigma, e1, e2, arguments, atmSummary, atmKwargs) = tuple(data.values())
    corrs_4m,_ = corr1d(thx,thy,sigma)
    corre1_4m,_ = corr1d(thx,thy,e1)
    corre2_4m,_ = corr1d(thx,thy,e2)
    corr4m[i] = [np.mean(corrs_4m), np.mean(corre1_4m), np.mean(corre2_4m)]
    
h2 = pd.DataFrame(corr2m, index=["sigma", "e1", "e2"]).T
h4 = pd.DataFrame(corr4m, index=["sigma", "e1", "e2"]).T

fig, ax = plt.subplots(2,3,figsize=(13,6))
ax[0,0].scatter(meanspd, h2['sigma'])
ax[0,1].scatter(meanspd, h2['e1'])
ax[0,2].scatter(meanspd, h2['e2'])

ax[1,0].scatter(meanspd, h4['sigma'])
ax[1,1].scatter(meanspd, h4['e1'])
ax[1,2].scatter(meanspd, h4['e2'])

[a.set_xlabel("mean speed") for a in ax[1]]
[ax[i,0].set_ylabel("mean $\\xi_{\sigma}$") for i in range(2)]
[ax[i,1].set_ylabel("mean $\\xi_{e1}$") for i in range(2)]
[ax[i,2].set_ylabel("mean $\\xi_{e2}$") for i in range(2)]
plt.tight_layout()    
```
